```python
import numpy as np

# Data initialization
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Hyperparameters and initial parameters
w = 0  # weight
b = 0  # bias
lr = 0.01  # learning rate
epochs = 50

# Training loop
for epoch in range(epochs):
    for i in range(len(x)):
        # Prediction (Forward pass)
        y_pred = w * x[i] + b

        # Calculate gradients (Backward pass)
        # Based on Mean Squared Error derivative
        dw = -2 * x[i] * (y[i] - y_pred)
        db = -2 * (y[i] - y_pred)

        # Update parameters
        w = w - lr * dw
        b = b - lr * db

# Final results
print("weight :", w)
print("Bias :", b)
```
