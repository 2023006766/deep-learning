```python
import numpy as np

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

def cross_entropy(y_true, y_pred):
    return -np.sum(y_true * np.log(y_pred + 1e-12))

# Weight matrices
Wxh = np.array([
[-1.753416, 0.9836007, 1.7042459822],
[-1.117427, 2.0291977, 0.5590203]
])

Whh = np.array([
[0.3323058, 1.704173],
[-0.4059835, 1.069699]
])

Why = np.array([
[5.0668307, 1.239699],
[-3.2882083, -2.307849]
])

# Biases
bh = np.array([
[-0.0238333],
[0.1707909]
])

by = np.array([
[0.03411441],
[-0.21828305]
])

# Inputs
x1 = np.array([[1],[0],[0]])
x2 = np.array([[0],[1],[0]])
x3 = np.array([[0],[0],[1]])

inputs = [x1, x2, x3]

# Targets
t1 = np.array([[0],[1]])
t2 = np.array([[1],[0]])
t3 = np.array([[0],[1]])

targets = [t1, t2, t3]

# Initial hidden state
h0 = np.array([[0],[0]])

hs = [h0]
ys = []
ps = []

loss = 0

for t in range(3):

    xt = inputs[t]

    wxhxt = Wxh @ xt
    print("wxhxt")
    print(wxhxt)

    whhhs = Whh @ hs[t]
    print("whhhs")
    print(whhhs)

    combined = wxhxt + whhhs + bh
    print("combined")
    print(combined)

    ht = np.tanh(combined)
    print("ht")
    print(ht)

    y = Why @ ht + by
    print("y")
    print(y)

    p = softmax(y)
    print("p")
    print(p)

    hs.append(ht)
    ys.append(y)
    ps.append(p)

    print("Current prediction (input):")
    print(xt)

    print("Next prediction (probabilities):")
    print(np.round(p,1))

    loss_entropy = cross_entropy(targets[t], p)
    print("loss-entropy")
    print(loss_entropy)

    loss += loss_entropy

print("Total Loss:", loss)

print("Hidden states:")
for i, h in enumerate(hs):
    print(f"h{i}:")
    print(h)
```
