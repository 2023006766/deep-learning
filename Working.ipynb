```python
import numpy as np

print("Working with Arrays")
print("=" * 70)

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

print(f"\nArray 1 (2x2):\n{arr1}")
print(f"Array 2 (2x2):\n{arr2}")

arr3 = np.arange(12).reshape(3, 4)
print(f"\nArray 3 (3x4):\n{arr3}")

print("\nArithmetic Operations")
print("-" * 50)
print(f"Addition:\n{arr1 + arr2}")
print(f"Subtraction:\n{arr1 - arr2}")
print(f"Multiplication:\n{arr1 * arr2}")
print(f"Division:\n{arr1 / arr2}")
print(f"Power:\n{arr1 ** 2}")

print("\nArray Properties")
print("-" * 50)
print(f"Shape arr1: {arr1.shape}")
print(f"Shape arr3: {arr3.shape}")
print(f"Dtype arr1: {arr1.dtype}")
print(f"Size arr1: {arr1.size}")
print(f"Dimensions arr1: {arr1.ndim}")

print("\nAggregate Functions")
print("-" * 50)
print(f"Sum: {np.sum(arr3)}")
print(f"Mean: {np.mean(arr3):.2f}")
print(f"Min / Max: {np.min(arr3)}, {np.max(arr3)}")
print(f"Standard Deviation: {np.std(arr3):.2f}")
print(f"Row sums: {np.sum(arr3, axis=1)}")
print(f"Column sums: {np.sum(arr3, axis=0)}")

print("\nArray Manipulation")
print("-" * 50)
print(f"Reshape (2x6):\n{np.arange(12).reshape(2, 6)}")
print(f"Flatten: {arr3.ravel()}")
print(f"Transpose:\n{arr3.T}")

print("\nUniversal Functions")
print("-" * 50)
print(f"Square Root:\n{np.sqrt(arr1)}")
print(f"Exponential (first row):\n{np.exp(arr1[:1])}")
print(f"Sine (arr3 values):\n{np.sin(arr3)}")

print("\nMatrix Operations")
print("-" * 50)
print(f"Dot product:\n{np.dot(arr1, arr2)}")
print(f"Matrix multiplication:\n{np.matmul(arr1, arr2)}")

print("\nStatistical Operations")
print("-" * 50)
print(f"Median: {np.median(arr3):.2f}")
print(f"25th Percentile: {np.percentile(arr3, 25):.2f}")

print("=" * 70)
```
