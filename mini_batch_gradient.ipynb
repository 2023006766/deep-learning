```python
import numpy as np

# Data initialization
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Hyperparameters and initial parameters
w = 0
b = 0
lr = 0.01
epochs = 50
batch_size = 2

# Training loop
for epoch in range(epochs):
    # Iterate through the data in steps of batch_size
    for i in range(0, len(x), batch_size):
        # Create mini-batches
        x_batch = x[i : i + batch_size]
        y_batch = y[i : i + batch_size]

        # Prediction for the batch
        y_pred = w * x_batch + b

        # Calculate gradients using the mean of the batch
        dw = -2 * np.mean(x_batch * (y_batch - y_pred))
        db = -2 * np.mean(y_batch - y_pred)

        # Update parameters
        w = w - lr * dw
        b = b - lr * db

# Final results
print("weight :", w)
print("Bias :", b)
```
