```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical

# 1. Loading the dataset and performing splitting
(train_x, train_y), (test_x, test_y) = mnist.load_data()

# 2. Reshaping and Normalization
train_x = train_x.reshape(train_x.shape[0], 28, 28, 1) / 255.0
test_x = test_x.reshape(test_x.shape[0], 28, 28, 1) / 255.0

# 3. One-hot encoding labels
train_y = to_categorical(train_y)
test_y = to_categorical(test_y)

# 4. Building the AlexNet Architecture
model = Sequential()

# Layer 1: Conv + MaxPool
model.add(Conv2D(96, (5, 5), strides=4, padding='valid', input_shape=(28, 28, 1), activation='relu'))
model.add(MaxPooling2D((2, 2), strides=2))

# Layer 2: Conv + MaxPool
model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D((2, 2), strides=1))

# Layer 3, 4, 5: Convolutions
model.add(Conv2D(384, (1, 1), padding='same', activation='relu'))
model.add(Conv2D(384, (1, 1), padding='same', activation='relu'))
model.add(Conv2D(256, (1, 1), padding='same', activation='relu'))

# Fully Connected Layers
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(10, activation='softmax'))

# 5. Compile the Model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# 6. Training (From your handwritten fit call)
model.fit(train_x, train_y,
          batch_size=128,
          epochs=10, https://storage.googleapis.com/tensorflow/tf-keras-datasets/mnist.npz
          verbose=1,
          validation_data=(test_x, test_y))

# 7. Evaluation
score = model.evaluate(test_x, test_y)
print('Test Loss:', score[0])
print('Test Accuracy:', score[1])
```
